# ASSIGNMENT 1:- CHI-SQUARE TEST

## Association between Device Type and Customer Satisfaction
**Background:**
Mizzare Corporation has collected data on customer satisfaction levels for two types of smart home devices: Smart Thermostats and Smart Lights. They want to determine if there's a significant association between the type of device purchased and the customer's satisfaction level.
Data Provided:
The data is summarized in a contingency table showing the counts of customers in each satisfaction level for both types of devices: 


# Assignment Tasks:

# Task:1)State the Hypotheses:

```python
#H0=there is no association between the type of smart home device purchased (Smart Thermostats vs. Smart Lights) and the customer satisfaction level.
# Ha=association between the type of smart home device purchased (Smart Thermostats vs. Smart Lights) and the customer satisfaction level.
```

```python
import pandas as pd
import numpy as np
import scipy.stats as stats
```

# Task:2)Compute the Chi-Square Statistic:

```python
from scipy.stats import chi2_contingency
data=np.array([[50,70],
        [80,100],
        [60,90],
        [30,50],
        [20,50]])
chi2_statistic,pvalue,dof,expected_freq =chi2_contingency(data)
print('chi2_statistic=',chi2_statistic),
print('pvalue=',pvalue),
print('Dof=',dof),
print('Expected_freq=',[expected_freq])
```

# Task:3)Determine the Critical Value:

```python
#a=0.5,CI=0.95
no_of_rows=5
no_of_columns=2
dof=(no_of_rows-1)*(no_of_columns-1)
dof
critical_value=stats.chi2.ppf(0.95,dof)
critical_value
```

# Task:4) Make a Decision:

```python
# Here,chi2_statistic<critical_value i.e;{5.638<9.487}
# failed to reject h0
```

# Conclusion:
AS we know already that chi2_statistic<critical_value, so I conclude that there is no association between the type of smart home device purchased (Smart Thermostats vs. Smart Lights) and the customer satisfaction level.

# ASSIGNMENT 2:- HYPOTHESIS TESTING

## Weekly Operating Cost for Restaurants
**Background:**
Bombay hospitality Ltd. operates a franchise model where weekly operating cost is modeled as:
$$ W = 1000 + 5X $$
where X is the number of units produced.

Owners claim that actual weekly costs are higher than the model.

**Data Provided:**
- Theoretical model: W = 1000 + 5X
- For X = 600 units → W = 1000 + 5*600 = 4000
- Sample mean weekly cost = 3050
- Sample size (n) = 25
- Units distribution: mean = 600, std dev = 25
- σ = 5 * 25 = 125

```python
#H0:It is not true that the restaurant owners' claim that the weekly operating costs are higher than the model suggests.
#Ha:restaurant owner's claim that the weekly operating costs are higher than the model suggests.
```
